### Chemical salinity test (NaCl)

```python
qp.quantif("../results/paper_test/salinity_chemical.csv",['0',"4.4",'11','22','44','NC-0','NC-44'])
plt.xlabel("Salt concentration of NaCl solution (g/L)")
plt.tight_layout()
plt.savefig(path+'salinity_chemical.png')
```

```python
m = qp.quantif("../results/paper_test/salinity_chemical.csv",['0',"4.4",'11','22','44','NC_0','NC_100'])[0]
plt.close()

sns.lineplot([x for x in range(len(m))],m,linestyle='--',marker='o',markeredgecolor='black',markerfacecolor='black')
plt.xticks([x for x in range(len(m))],['0',"4.4",'11','22','44','NC_0','NC_100'])
plt.xlabel("Salt concentration of NaCl solution (g/L)")
plt.ylabel(r'$ \frac{test}{control+test} $')
plt.savefig(path+'salinity_chemical_line.png')
plt.show()
```
